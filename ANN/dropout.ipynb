```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=500, n_features=5, noise=40)
```

## WithOut DropOut

```python
model = Sequential()

model.add(Dense(8,activation='relu',input_shape=(5,)))
model.add(Dense(8,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.compile(optimizer=SGD(learning_rate=0.01),loss='mse',metrics=['mse'])
```

```python
history = model.fit(X,y,epochs=30,verbose=1,validation_split=0.2,batch_size=10)
```

```python
plt.plot(history.history['loss'],label='train_loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

## Uisng DropOut

```python
from tensorflow.keras.layers import Dropout
```

```python
model_drop = Sequential()
```

```python
model_drop.add(Dense(8,activation='relu',input_shape=(5,)))
model_drop.add(Dropout(0.20))
model_drop.add(Dense(8,activation='relu'))
model_drop.add(Dropout(0.20))
model_drop.add(Dense(4,activation='relu'))
model_drop.add(Dropout(0.10))
model_drop.add(Dense(4,activation='relu'))
model_drop.add(Dense(2,activation='relu'))
model_drop.add(Dense(1,activation='linear'))
```

```python
model_drop.compile(optimizer=SGD(learning_rate=0.01),loss='mse',metrics=['mse'])
```

```python
history = model_drop.fit(X,y,epochs=30,verbose=1,validation_split=0.2,batch_size=10)
```

```python
plt.plot(history.history['loss'],label='train_loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```
